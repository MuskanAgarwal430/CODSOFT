## Build Model

```python
X = df.drop( 'species', axis=1)
y = df['species']
print(X.head())
print(X.shape)
print(y.head())
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=10)
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

## 1. Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
logr = LogisticRegression(max_iter=1000)  # Increase the number of iterations
logr.fit(X_train,y_train)
y_pred = logr.predict(X_test)
acc_log = metrics.accuracy_score(y_pred,y_test)
print('The accuracy of the Logistic Regression is', acc_log)
```

## 2. Support Vector Machine (SVM)

```python
from sklearn.svm import SVC
sv = SVC()
sv.fit(X_train,y_train)
y_pred = sv.predict(X_test)
acc_svm = metrics.accuracy_score(y_pred,y_test)
print('The accuracy of the SVM is:', acc_svm)
```

## 3. Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
nb_classifier = GaussianNB()
nb_classifier.fit(X_train, y_train)
y_pred = nb_classifier.predict(X_test)
acc_nb = metrics.accuracy_score(y_test, y_pred)
print(f'Accuracy: {acc_nb:.2f}')
```

## 4. K-Nearest Neighbours

```python
from sklearn.neighbors import KNeighborsClassifier
knc = KNeighborsClassifier(n_neighbors=3)
knc.fit(X_train,y_train)
y_pred = knc.predict(X_test)
acc_knn = metrics.accuracy_score(y_pred,y_test)
print('The accuracy of the KNN is', acc_knn)
```

## 5. Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(X_train,y_train)
y_pred = dt.predict(X_test)
acc_dt = metrics.accuracy_score(y_pred,y_test)
print('The accuracy of the Decision Tree is', acc_dt)
```

```python
a_index = list(range(1, 11))
a = []
for i in a_index:
    kcs = KNeighborsClassifier(n_neighbors=i)
    kcs.fit(X_train, y_train)
    y_pred = kcs.predict(X_test)
    accuracy = metrics.accuracy_score(y_pred, y_test)
    a.append(accuracy)
plt.plot(a_index, a, marker='o')
plt.xlabel('Number of Neighbors (k)')
plt.ylabel('Accuracy')
plt.title('K-Nearest Neighbors Accuracy for Different k Values')
plt.xticks(a_index)
plt.grid(True)
plt.show()
```

## 6. Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)
rf_classifier.fit(X_train, y_train)
y_pred = rf_classifier.predict(X_test)
acc_rm = accuracy_score(y_test, y_pred)
print(f'The accuracy of the Random Forest is {accuracy:.2f}')
```

```python
print(classification_report(y_test, y_pred))
print('Confusion Matrix:')
print(confusion_matrix(y_test, y_pred))
```
